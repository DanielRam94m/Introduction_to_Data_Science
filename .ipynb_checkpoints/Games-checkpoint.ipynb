```python
df.groupby(["Genre"]).mean()
```

#### Printing Mean of an specific //without resizing

```python
saleMeans = df.groupby(["Genre"]).mean()
plt.plot(saleMeans["NA_Sales"])
```

#### Chart Resizing & Exporting it as an image

```python
saleMeans = df.groupby(["Genre"]).mean()
plt.figure(figsize=(12, 6))
plt.plot(saleMeans["NA_Sales"])

plt.savefig("Sales_na.jpg")
```

#### Show two different averages in contrast

```python
meanComparison = df.groupby(['Genre']).mean()
plt.figure(figsize=(12,6))

sales_na = meanComparison['NA_Sales']
sales_eu = meanComparison['EU_Sales']

plt.plot(sales_na, color="red")
plt.plot(sales_eu, color="purple")
```
